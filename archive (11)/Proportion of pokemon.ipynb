```python
df['Type 1'].value_counts()
```

```python
type1_counts = df['Type 1'].value_counts()
```

```python
plt.figure(figsize=(8, 10))
plt.pie(type1_counts, labels=type1_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Type 1')
plt.axis('equal') 
plt.show()
```

```python
df['Generation'].value_counts()
```

```python
generation_counts = df['Generation'].value_counts()
```

```python
plt.figure(figsize=(8, 10))
plt.pie(generation_counts, labels=generation_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Generations')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

### Over All pokemon Stats Max

```python
# Find the maximum HP value and the corresponding Pokémon
max_hp = df['HP'].max()
max_hp_pokemon = df[df['HP'] == max_hp]

print("Maximum HP value:", max_hp)
print("Pokémon with maximum HP:")
print(max_hp_pokemon[['Name', 'HP']])

# Find the maximum Attack value and the corresponding Pokémon
max_attack = df['Attack'].max()
max_attack_pokemon = df[df['Attack'] == max_attack]

print("Maximum attack value:", max_attack)
print("Pokémon with maximum Attack:")
print(max_attack_pokemon[['Name', 'Attack']])  

# Find the maximum defence value and the corresponding Pokémon
max_Defense = df['Defense'].max()
max_Defense_pokemon = df[df['Defense'] == max_attack]

print("Maximum Defense value:", max_Defense)
print("Pokémon with maximum Defense:")
print(max_attack_pokemon[['Name', 'Defense']]) 

# Find the maximum Sp Atk value and the corresponding Pokémon
max_SpAtk = df['Sp. Atk'].max()
max_SpAtk_pokemon = df[df['Sp. Atk'] == max_SpAtk]

print("Maximum Sp. Atk value:", max_SpAtk)
print("Pokémon with maximum Sp. Atk:")
print(max_attack_pokemon[['Name', 'Sp. Atk']]) 

# Find the maximum Sp. Def value and the corresponding Pokémon
max_SpDef = df['Sp. Def'].max()
max_SpDef_pokemon = df[df['Sp. Def'] == max_SpDef]

print("Maximum Sp. Def:", max_SpDef)
print("Pokémon with maximum Sp. Def:")
print(max_attack_pokemon[['Name', 'Sp. Def']]) 

# Find the maximum Speed value and the corresponding Pokémon
max_Speed = df['Speed'].max()
max_Speed_pokemon = df[df['Speed'] == max_Speed]

print("Maximum Speed:", max_Speed)
print("Pokémon with maximum Speed:")
print(max_Speed_pokemon[['Name', 'Speed']]) 
```
